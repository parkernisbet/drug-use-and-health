```python
## loop through years 2002 to 2018, pull value_counts for each drug
import pandas as pd

# ages surveyed, detailed in dataset metadata
ages = range(1, 101)
# age columns in data sets
targets = ['CIGTRY', 'ALCTRY', 'MJAGE', 'COCAGE', 'CRKAGE', 'HERAGE', 'HALLAGE', 'INHAGE', 'ANALAGE', 'TRANAGE', 'STIMAGE', 'SEDAGE']
# exception age columns for years 2016 and 2018
exceptions = {'HALLAGE': 'HALLUCAGE', 'INHAGE': 'INHALAGE', 'ANALAGE': 'PNRNMAGE', 'TRANAGE': 'TRQNMAGE', 'STIMAGE': 'STMNMAGE', 'SEDAGE': 'SEDNMAGE'}
for i in range(2002, 2019):
    year = i
    exec('df_drugAge{} = pd.DataFrame(data = ages, columns = [\'Age\']).set_index(\'Age\')'.format(str(i)))
    # construct filename and read into dataframe
    filename = 'NSDUH_' + str(year) + '_Tab.tsv'
    df_drugs = pd.read_csv(filename, sep = '\t')
    for col in targets:
        # exception usage
        if i in [2015, 2016, 2017, 2018] and col in exceptions.keys():
            colException = exceptions[col]
        else:
            colException = col
        # store value_counts
        dict_counts = dict(df_drugs[colException].value_counts())
        for j, k in dict_counts.items():
            exec('df_drugAge{}.loc[j, col] = int(k)'.format(str(i)))
    # replace nans with 0
    exec('df_drugAge{} = df_drugAge{}.fillna(0)'.format(str(i), str(i)))

df_drugAge2002.head()
```

```python
## segment dataframes by drug type
drugs = {'tobacco': 'CIGTRY', 'alcohol': 'ALCTRY', 'marijuana': 'MJAGE', 'cocaine': 'COCAGE', 'crack': 'CRKAGE', 'heroin': 'HERAGE', 'hallucinogens': 'HALLAGE', 
         'inhalants': 'INHAGE', 'painRelievers': 'ANALAGE', 'tranquilizers': 'TRANAGE', 'stimulants': 'STIMAGE', 'sedatives': 'SEDAGE'}
for word in drugs.keys():
    exec('df_{} = pd.DataFrame(data = ages, columns = [\'Age\']).set_index(\'Age\')'.format(word))
    for i in range(2002, 2019):
        exec('df_{}[{}] = df_drugAge{}[drugs[word]][:100]'.format(word, str(i), str(i)))
```

```python
## save dataframes as excel files
for i in range(2002, 2019):
    exec('df_drugAge{}.to_excel(\'df_drugAge{}.xlsx\')'.format(str(i), str(i)))

for word in drugs.keys():
    exec('df_{}.to_excel(\'df_{}.xlsx\')'.format(word, word))
```
